# 뉴노멀 MBA: Product를 팔리게 하는, 소비자 행동과 마케팅 전략
- 마켓 1.0 시대엔 제품 공급이 고객의 수요를 따라잡지 못하는 기술력을 가졌던 때로 물건을 만들기만 하면 재고 없이 팔리던 시대로 성능 홍보, 기술력 향상에 집중했다.
- 기술 발달과 고객들에게 다양한 선택지, 니즈가 생기며 모든 소비자를 만족시키는 제품 생산은 불가능해졌다.
- 자사가 성공적으로 공략할 수 있는 세분 시장을 정의하고 역량을 집중하는 `STP 전략`(세분화 전략)을 선보이고 기업이 의도한 생산, 가격 및 전달 등을 위한 구체적인 마케팅 믹스를 정한다.
- 인터넷, SNS 등의 발달로 고객 경로는 광고에 기반한 일방향적인 형태에서 브랜드, 타소비자와의 소통까지 다양해졌다.
- 스토리텔링을 통한 홍보가 중요하다고 언급했으나 개인적으론 최근 '15초의 광고시간도 길다, 6초 광고가 효과적'이라는 조사 결과가 많은 만큼 스토리텔링을 활용해 메세지를 전달하긴 쉽지 않아 보인다. <br>
  [기사] https://www.asiae.co.kr/news/view.htm?idxno=2018040315370683523
- 자사 옹호 고객 확보가 중요하다. 단 모델(인플루언서 등), 제품에 문제가 있을 경우 옹호 고객의 모델간 애착, 브랜드 호감도 등이 크게 하락해 더 큰 노이즈에 휘말릴 수 있다.

- 브랜드 충성도 구매(고관여 제품, 반복 구매)는 기업이 확보해야 할 구매행동으로 고관여제품을 브랜드 신뢰, 이미지로 자주 구매하는 것을 뜻한다.
- 다양성 추구 구매(저관여 제품, 최초 구매)는 관성적 구매를 하다 질리거나 새로운 시도를 하기 위해 이루어진 것으로 본다.
- 관성적 구매(저관여 제품, 반복 구매)는 인지적 수고를 덜기 위해 반복된 구매를 하는 것으로 타사에서 세일을 한다면 얼마든지 옮겨갈 수 있다.
- 맥락 효과, 감정적 소비가 늘어나고 있다.
- 소비자 구매행동은 구매결정에서 끝나지 않고 사용하며 느낀 경험을 공유하는 것까지 포함된다. 혹평을 받아도 고객이 건의한 문제를 진정성있게 해결한다면 불평을 갖지 않은 고객보다도 재이용률을 높일 수 있다.(Service Recovery Paradox)

### 소비자 정보 처리
- 많은 정보를 제공하지 않아 궁금증을 유발하는 티저 광고를 이용하는 것은 제품 노출에 유용하다.
- 시각적 요소에 생기는 지각적 조직화, 불완전한 광고로 성과 향상
- 개인별 지각적 범주화, 가격&품질 연상 등에 의한 지각적 추론에 따라 다르게 해석됨
